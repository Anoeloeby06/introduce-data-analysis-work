```python
# import library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['font.size'] = 14
plt.rcParams['font.weight'] = 'bold'
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
path = '../input/'

df = pd.read_csv('Heart_Disease_Prediction.csv')

print(f'\nNumber of rows and columns in the dataset:\n{df.shape}\n')
print('')
df.head()
```

```python
sns.lmplot(x='BP', y='Max HR', data=df, aspect=2, height=6)
plt.xlabel('Blood Pressure$(mmHg)$: as Independent Variable')
plt.ylabel('Max Heart Rate as Dependent Variable')
plt.title('BP VS Max HR')
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,4))
sns.heatmap(df.isnull(),cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Value in the Dataset')
```

```python
#correlation plot
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='viridis', annot=True);
```

```python
# Distribution Plot
f = plt.figure(figsize=(10,4))

ax =f.add_subplot(121)
sns.distplot(df['BP'], bins=50, color='r', ax=ax)
ax.set_title('Distribution of Blood Pressure')

ax =f.add_subplot(122)
sns.distplot(np.log10(df['BP']), bins=40, color='y', ax=ax)
ax.set_title('Distribution of Blood Pressure on $log$ scale')
ax.set_xscale('log');
```

```python
# Violin Plot
f = plt.figure(figsize=(10,4))

ax =f.add_subplot(121)
sns.violinplot(data=df, x='Chest pain type',y='Cholesterol', palette='Wistia', ax=ax)
ax.set_title('Violin Plot of Chest Pain Type vs Cholesterol')

ax =f.add_subplot(122)
sns.violinplot(data=df, x='Heart Disease',y='Age', palette='magma', ax=ax)
ax.set_title('Violin Plot of Heart Disease vs Age');
```

```python
# Box Plot
sns.boxplot(data=df, x='Chest pain type',y='BP', palette='Wistia', hue='Sex')
ax.set_title('Box Plot of Chest Pain Type vs Blood Pleasure')
```

```python
df.groupby('Chest pain type').agg(['mean', 'min', 'max'])['Max HR'];
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Heart Disease',y='Cholesterol', hue='Sex',palette='rainbow', split=True)
ax.set_title('Violin Plot of Heart Disease vs Cholesterol')
```

```python
f = plt.figure(figsize=(12,6))
ax = f.add_subplot(121)
sns.scatterplot(data=df, x='Cholesterol',y='Max HR', hue='Heart Disease',palette='magma', ax=ax)
ax.set_title('Scatter Plot of Max HR vs Cholesterol')

ax =f.add_subplot(122)
sns.scatterplot(data=df, x='BP',y='Max HR', hue='Heart Disease',palette='viridis', ax=ax)
ax.set_title('Scatter Plot of Max HR vs BP');

plt.savefig('Sc.png')
```
